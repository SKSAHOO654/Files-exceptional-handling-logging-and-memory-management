1 What is the difference between interpreted and compiled languages ?

- Interpreted Languages
  - Code is executed line by line by an interpreter at runtime.
  - No separate compilation step; execution happens directly from source code.
  - Slower than compiled languages since interpretation happens at runtime.
  - Easier for debugging because errors are detected as the code runs.
    - Examples: Python, JavaScript, Ruby, PHP

- Compiled Languages
 - Code is translated entirely into machine code before execution.
 - Requires a compiler to convert source code into an executable file.
 - Faster execution since machine code runs directly on hardware.
 - More efficient but harder to debug because errors appear at compile time.
   - Examples: C, C++, Rust, Go

9 What is the difference between multithreading and multiprocessing ?


1. Multithreading
- Uses multiple threads within a single process.
- Threads share the same memory space, making communication easier but leading to potential race conditions.
- Limited by the Global Interpreter Lock (GIL) in Python, meaning only one thread executes Python bytecode at a time.
- Best for I/O-bound tasks (e.g., file I/O, network requests).

2. Multiprocessing
- Uses multiple processes, each with its own memory space.
- Bypasses the GIL, allowing true parallel execution.
- Best for CPU-bound tasks (e.g., heavy computations, data processing).

18 What is the difference between os.fork() and multiprocessing in Python ?


1. os.fork(): Low-Level Process Creation
- os.fork() is a low-level system call used to create a child process by duplicating the current process.
- Available only on UNIX-based systems (Linux, macOS).
- The child process gets a copy of the parent’s memory space.
- Must check the return value to determine if it’s the parent (PID > 0) or child (PID = 0).

2. multiprocessing: High-Level Process Management
- multiprocessing is a cross-platform module for creating multiple processes.
- Works on Windows, Linux, macOS.
- Uses Process objects instead of direct system calls.
- Handles memory, synchronization, and inter-process communication (IPC) automatically.



19 What is the importance of closing a file in Python ?


Importance of Closing a File in Python
- When working with files in Python, it is crucial to close the file after use. Failing to do so can lead to data corruption, memory leaks, and resource wastage.

1. Frees Up System Resources
When a file is open, it consumes system resources (like memory and file handles).
 - Closing the file ensures efficient resource management.
2. Ensures Data is Written to Disk
When writing to a file, data may be buffered (temporarily stored in memory).
 - Closing the file flushes the buffer, ensuring all data is saved properly.
3. Prevents Data Corruption
If a file remains open, especially for writing, and the program crashes, data may be lost or corrupted.
 - Closing the file ensures the file is properly updated and safe.
4. Avoids Too Many Open Files Error
Opening multiple files without closing them can lead to an "Too many open files" error.
 - This happens because the OS limits the number of open files per process.

23 What are the challenges associated with memory management in Python ?


1. Garbage Collection Overhead
- Python uses automatic garbage collection (GC) to free up memory, but:
  - The GC runs periodically, not immediately.
  - It can pause execution temporarily, affecting performance.
  - Circular references may not be collected immediately.

2. Reference Counting Issues
- Python keeps track of references to an object.
 -If an object’s reference count drops to zero, it gets deleted automatically.
- Problem: Circular references (where two objects reference each other) prevent memory from being

3. Memory Leaks
- If objects are referenced indefinitely, Python cannot free the memory.
Common causes:
  - Global variables holding references to objects.
 - Unclosed file handles or database connections.
 - Improper use of caches or lists storing large objects.

4. High Memory Usage in Large Data Processing
- When handling large datasets, Python may consume too much RAM.
 - Example: Processing large lists or NumPy arrays in memory.
- Since Python is dynamically typed, it uses extra memory for object storage
- compared to lower-level languages like C.

5. Python’s Global Interpreter Lock (GIL)
- The GIL ensures that only one thread executes Python bytecode at a time.
- This means multithreading won’t help in CPU-intensive tasks, limiting memory efficiency.

6. Fragmentation Issues
- Python’s memory allocator doesn’t always release memory back to the OS.
- Memory fragmentation occurs when objects of different sizes are allocated and freed, leading to inefficient space usage.

7. Objects Staying in Memory Too Long
- Objects stored in caches, global variables, or modules may not be freed.
- Example: Keeping a large Pandas DataFrame in memory unnecessarily.
